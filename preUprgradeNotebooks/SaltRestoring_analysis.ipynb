```python
x="/scratch/general/am8e13/smallSalt36km/"
#x = "/hpcdata/scratch/am8e13/cs_36km_tutorial/run_SaltIni/results"
os.chdir(x)                                                                                                                                                                                                                  
file2read = netcdf.NetCDFFile("state.nc",'r')
Y=file2read.variables['Y']
Y=Y[:]*1
S_pert=file2read.variables['S']
S_pert=S_pert[:]*1
Z=file2read.variables['Z']
Z=Z[:]*1
X=file2read.variables['X']
X=X[:]*1
I=file2read.variables['iter']
I=I[:]*1
days=file2read.variables['T']
days=days[:]*1

#file2read = netcdf.NetCDFFile("salt.nc",'r')
#S_pert_big=file2read.variables['S']
#S_pert_big=S_pert_big[:]*1

x="/scratch/general/am8e13/small36km/"
os.chdir(x)                                                                                                                                                                                                                  
file2read = netcdf.NetCDFFile("state.nc",'r')
S_np=file2read.variables['S']
S_np=S_np[:]*1
```

```python
dx=36000
t = 60*60*24*360 #seconds in an year
```

```python
# Bathy is 1 on land and 0 over sea
x="/scratch/general/am8e13/results36km"
os.chdir(x) 
file2read = netcdf.NetCDFFile("grid.nc",'r')
Depth=file2read.variables['Depth']
Depth=Depth[:]*1
bathy=file2read.variables['HFacC']
bathy=bathy[:]*1
bathy[bathy==0]=-1
bathy[bathy>0]=0
bathy[bathy==-1]=1

bathy1 = bathy[0,:,:]
```

```python
# Tracer analysis 
x="/scratch/general/am8e13/smallSalt36km/"
os.chdir(x)
file2read = netcdf.NetCDFFile("ptracers.nc",'r')
# Tracers
tr1=file2read.variables['tr1']
tr1=tr1[:]*1
#tr2=file2read.variables['tr2']
#tr2=tr2[:]*1

tr1_top_all = sum(tr1,axis = 1)
tr1_top_small = sum(tr1[:,35:45,:,:],axis = 1)
tr1_top_up = sum(tr1[:,0:35,:,:],axis = 1)
```

```python
# Sediment spreading in the storegga waterdepth

fig, axes = plt.subplots(nrows=2, ncols=3)
i=0
bathyfig = bathy1*tr1_top_small.max()*2

vimin = 0
vimax = tr1_top_small.max()*2
#levels = linspace(vimin, vimax, num=10)
#print levels
for ax in axes.flat:
    cs = ax.contourf(X,Y,tr1_top_small[i*120,:,:]+ bathyfig,100,vmin=vimin, vmax=vimax)
    fig.colorbar(cs, ax=ax)
    ax.set_title(""+str((days[(i*120)]-days[0])//t) +" years")
    i=i+1
#
fig.subplots_adjust(right=2.4,top=2)
```

```python
# Sediment spreading above the Storegga layer

fig, axes = plt.subplots(nrows=2, ncols=3)
i=0
bathyfig = bathy1*tr1_top_up.max()*2

vimin = 0
vimax = tr1_top_up.max()*2
#levels = linspace(vimin, vimax, num=10)
#print levels
for ax in axes.flat:
    cs = ax.contourf(X,Y,tr1_top_up[i*120,:,:]+ bathyfig,100,vmin=vimin, vmax=vimax)
    fig.colorbar(cs, ax=ax)
    ax.set_title(""+str((days[(i*120)]-days[0])//t) +" years")
    i=i+1
#
fig.subplots_adjust(right=2.4,top=2)
```

```python
# Sediment spreading above the Storegga layer

fig, axes = plt.subplots(nrows=2, ncols=3)
i=0
bathyfig = bathy1*tr1_top_all.max()*2

vimin = 0
vimax = tr1_top_all.max()*2
#levels = linspace(vimin, vimax, num=10)
#print levels
for ax in axes.flat:
    cs = ax.contourf(X,Y,tr1_top_all[i*120,:,:]+ bathyfig,100,vmin=vimin, vmax=vimax)
    fig.colorbar(cs, ax=ax)
    ax.set_title(""+str((days[(i*120)]-days[0])//t) +" years")
    i=i+1
#
fig.subplots_adjust(right=2.4,top=2)
```
