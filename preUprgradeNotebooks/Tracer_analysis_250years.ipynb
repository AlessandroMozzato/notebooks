```python
from scipy.io import netcdf
import numpy as np
from pylab import clf, plot, show, floor, ceil, imshow
import matplotlib.pyplot as plt 
#xlabel, ylabel, legend, savefig, colorbar, title, clim, pcolor, cm, contourf
import os
import csv
import sys
import glob
from IPython.display import display, Math, Latex
#sys.path.append('/noc/users/am8e13/Python/python_functions/')
#from Variable_button import *
```

```python
dx=36000
t = 60*60*24*360 #seconds in an year
```

```python
# Bathy is 1 on land and 0 over sea
x="/scratch/general/am8e13/results36km"
os.chdir(x) 
file2read = netcdf.NetCDFFile("grid.nc",'r')
Depth=file2read.variables['Depth']
Depth=Depth[:]*1
bathy=file2read.variables['HFacC']
bathy=bathy[:]*1
bathy[bathy==0]=-1
bathy[bathy>0]=0
bathy[bathy==-1]=1

bathy1 = bathy[0,:,:]
```

```python
# Tracer analysis 
x="/scratch/general/am8e13/resultsSalt36km/"
os.chdir(x)
file2read = netcdf.NetCDFFile("ptracers.nc",'r')
# Tracers
tr1=file2read.variables['tr1']
tr1=tr1[:]*1
tr2=file2read.variables['tr2']
tr2=tr2[:]*1

Y=file2read.variables['Y']
Y=Y[:]*1
Z=file2read.variables['Z']
Z=Z[:]*1
X=file2read.variables['X']
X=X[:]*1
I=file2read.variables['iter']
I=I[:]*1
days=file2read.variables['T']
days=days[:]*1

tr1_top_all = sum(tr2,axis = 1)
tr1_top_small = sum(tr2[:,35:45,:,:],axis = 1)
tr1_top_up = sum(tr2[:,0:35,:,:],axis = 1)
```

```python
# Sediment spreading in the storegga waterdepth

fig, axes = plt.subplots(nrows=2, ncols=3)
i=0
bathyfig = bathy1*tr1_top_small[100:,:,:].max()

vimin = 0
vimax = tr1_top_small[100:,:,:].max()
#levels = linspace(vimin, vimax, num=10)
#print levels
for ax in axes.flat:
    cs = ax.contourf(X,Y,tr1_top_small[i*300,:,:]+ bathyfig,100,vmin=vimin, vmax=vimax)
    fig.colorbar(cs, ax=ax)
    ax.set_title(""+str((days[(i*300)]-days[0])//t) +" years")
    i=i+1
#
fig.subplots_adjust(right=2.4,top=2)
```

```python
# Sediment spreading above the Storegga layer

fig, axes = plt.subplots(nrows=2, ncols=3)
i=0
bathyfig = bathy1*tr1_top_up[100:,:,:].max()

vimin = 0
vimax = tr1_top_up[100:,:,:].max()
#levels = linspace(vimin, vimax, num=10)
#print levels
for ax in axes.flat:
    cs = ax.contourf(X,Y,tr1_top_up[i*300,:,:]+ bathyfig,100,vmin=vimin, vmax=vimax)
    fig.colorbar(cs, ax=ax)
    ax.set_title(""+str((days[(i*300)]-days[0])//t) +" years")
    i=i+1
#
fig.subplots_adjust(right=2.4,top=2)
```

```python
# Sediment spreading on the whole water column

fig, axes = plt.subplots(nrows=2, ncols=3)
i=0
bathyfig = bathy1*tr1_top_all[100:,:,:].max()

vimin = 0
vimax = tr1_top_all[100:,:,:].max()
#levels = linspace(vimin, vimax, num=10)
#print levels
for ax in axes.flat:
    cs = ax.contourf(X,Y,tr1_top_all[i*100,:,:]+ bathyfig,100,vmin=vimin, vmax=vimax)
    fig.colorbar(cs, ax=ax)
    ax.set_title(""+str((days[(i*100)]-days[0])//t) +" years")
    i=i+1
#
fig.subplots_adjust(right=2.4,top=2)
```

```python
x="/scratch/general/am8e13/small36km/"
#x = "/hpcdata/scratch/am8e13/cs_36km_tutorial/run_SaltIni/results"
os.chdir(x)                                                                                                                                                                                                                  
file2read = netcdf.NetCDFFile("tracer1diag.nc",'r')
# Zonal Mass Weighed transport (kg/kg).m/s
UTRAC01=file2read.variables['UTRAC01']
UTRAC01=UTRAC01[:]*1 
# Merid Mass-Weighted Transp (kg/kg).m/s
VTRAC01=file2read.variables['VTRAC01']
VTRAC01=VTRAC01[:]*1
# Vertical Mass-Weighted Transp (kg/kg).m/s
WTRAC01=file2read.variables['WTRAC01']
WTRAC01=WTRAC01[:]*1
# Zonal Advective Flux (kg/kg).m^3/s
ADVxTr01=file2read.variables['ADVxTr01']
ADVxTr01=ADVxTr01[:]*1
# Meridional Advective Flux (kg/kg).m^3/s
ADVyTr01=file2read.variables['ADVyTr01']
ADVyTr01=ADVyTr01[:]*1
# Vertical Diffusive flux Explicit (kg/kg).m^3/s
DFrETr01=file2read.variables['DFrETr01']
DFrETr01=DFrETr01[:]*1
# Vertical Diffusive flus Implicit (kg/kg).m^3/s
DFrITr01=file2read.variables['DFrITr01']
DFrITr01=DFrITr01[:]*1
# Zonal Diffusive Flux (kg/kg).m^3/s
DFxETr01=file2read.variables['DFxETr01']
DFxETr01=DFxETr01[:]*1
# Meridional Diffusive Flux (kg/kg).m^3/s
DFyETr01=file2read.variables['DFyETr01']
DFyETr01=DFyETr01[:]*1
```

```python
UTRAC01_ave = sum(UTRAC01,axis = 1)
VTRAC01_ave = sum(VTRAC01,axis = 1)
WTRAC01_ave = sum(WTRAC01,axis = 1)

ADVxTr01_ave = sum(ADVxTr01,axis = 1)
ADVyTr01_ave = sum(ADVyTr01,axis = 1)

DFrETr01_ave = sum(DFrETr01,axis = 1)
DFrITr01_ave = sum(DFrITr01,axis = 1)

DFxETr01_ave = sum(DFxETr01,axis = 1)
DFyETr01_ave = sum(DFyETr01,axis = 1)

UTRAC01_ave_tot = sum(UTRAC01_ave,axis = 0)
VTRAC01_ave_tot = sum(VTRAC01_ave,axis = 0)
WTRAC01_ave_tot = sum(WTRAC01_ave,axis = 0)

ADVxTr01_ave_tot = sum(ADVxTr01_ave,axis = 0)
ADVyTr01_ave_tot = sum(ADVyTr01_ave,axis = 0)

DFrETr01_ave_tot = sum(DFrETr01_ave,axis = 0)
DFrITr01_ave_tot = sum(DFrITr01_ave,axis = 0)

DFxETr01_ave_tot = sum(DFxETr01_ave,axis = 0)
DFyETr01_ave_tot = sum(DFyETr01_ave,axis = 0)
```
